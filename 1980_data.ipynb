```python
import numpy as np
import pandas as pd
```

```python
data_1980 = pd.read_csv("data/1980_tracts.csv")
tract_puma = pd.read_csv("data/tract_puma.txt", dtype=str)
```

```python
data_1980
```

```python
tract_puma["PUMA"] = tract_puma["STATEFP"] + tract_puma["PUMA5CE"]
tract_puma["tract"] = tract_puma["STATEFP"] + tract_puma["COUNTYFP"] + tract_puma["TRACTCE"]
tract_puma = tract_puma.set_index("tract")
tract_puma.head()
```

```python
data_1980["Geo_FIPS"] = data_1980["Geo_FIPS"].astype(str).str.zfill(11)
data_1980
```

```python
data_1980["PUMA"] = tract_puma.loc[data_1980["Geo_FIPS"], "PUMA"].values
data_1980
```

```python
data_1980["HH_MED_OWN"] = (data_1980["HH_OWN_M"] * data_1980["HU_OWN_M"] + data_1980["HH_OWN_NOM"] * data_1980["HU_OWN_NOM"]) / (data_1980["HU_OWN_M"] + data_1980["HU_OWN_NOM"])
```

```python
data_1980["HHCOST"] = (data_1980["HH_MED_RENT"] * data_1980["HU_RENT"] + data_1980["HH_MED_OWN"] * data_1980["HU_OWN"]) / (data_1980["HU_RENT"] + data_1980["HU_OWN"])
```

```python
data_1980["HU_OWN_NOM"].isna().sum()
```

```python
data_1980[data_1980["PUMA"] == "0600101"][["HH_MED_OWN", "HH_MED_RENT", "HHCOST"]]
```

```python
data_1980["HH_MED_INC_adj"] = data_1980["HH_MED_INC"] * data_1980["TOT_POP"]
data_1980["HH_MED_OWN_adj"] = data_1980["HH_MED_OWN"] * data_1980["TOT_POP"]
data_1980["HH_COST_adj"] = data_1980["HHCOST"] * data_1980["TOT_POP"]
data_1980["HH_MED_RENT_adj"] = data_1980["HH_MED_RENT"] * data_1980["TOT_POP"]
```

```python
data_1980.columns
```

```python
puma_1980 = data_1980.groupby("PUMA").sum()[["HH_MED_INC_adj", "HH_MED_OWN_adj", "HH_MED_RENT_adj", "HH_COST_adj", "TOT_POP"]]
puma_1980["HH_MED_INC_adj"] = puma_1980["HH_MED_INC_adj"] / puma_1980["TOT_POP"]
puma_1980["HH_MED_RENT_adj"] = puma_1980["HH_MED_RENT_adj"] / puma_1980["TOT_POP"]
puma_1980["HH_MED_OWN_adj"] = puma_1980["HH_MED_OWN_adj"] / puma_1980["TOT_POP"]
puma_1980["HH_COST_adj"] = puma_1980["HH_COST_adj"] / puma_1980["TOT_POP"]
```
