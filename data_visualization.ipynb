```python
# 1 is owned with mortgage, 2 is owned free and clear, 3 is rented, 4 is occupied wdithout payment of rent
fig, ax = plt.subplots(figsize=(20, 7))
ct = pd.crosstab(df["TEN"], df["STAY"], normalize="index")
    
ct.plot(kind='bar', stacked=True, rot=0, mark_right=True, ax=ax)

ct.plot(kind='bar', stacked=True, rot=0, mark_right=True, ax=ax)
for n, x in enumerate([*ct.index.values]):
    for proportion in ct.loc[x]:
                
        plt.text(x=n,
                 y=proportion,
                 s=f'{np.round(proportion * 100, 1)}%', 
                 color="black",
                 fontsize=12,
                 fontweight="bold")
```

```python
df["SCHL_group"] = np.where(df["SCHL"] <= 15, 1, 0)
df["SCHL_group"] = np.where((df["SCHL"] <= 17) & (df["SCHL"] >= 16), 2, df["SCHL_group"])
df["SCHL_group"] = np.where((df["SCHL"] <= 20) & (df["SCHL"] >= 18), 3, df["SCHL_group"])
df["SCHL_group"] = np.where(df["SCHL"] >= 21, 4, df["SCHL_group"])
```

```python
df["SCHL_group"].value_counts()
```

```python
df["SCHL_group_group"] = np.where(df["SCHL_group"] == 4, 1, 0)
```

```python
# educational attainment; 1 is no high school, 2 is high school, 3 is some college, 4 is college+
fig, ax = plt.subplots(figsize=(10, 10))
ct = pd.crosstab(df["SCHL_group"], df["STAY"], normalize="index")
    
ct.plot(kind='bar', stacked=True, rot=0, mark_right=True, ax=ax)
for n, x in enumerate([*ct.index.values]):
    for proportion in ct.loc[x]:
                
        plt.text(x=n,
                 y=proportion,
                 s=f'{np.round(proportion * 100, 1)}%', 
                 color="black",
                 fontsize=12,
                 fontweight="bold")
```
